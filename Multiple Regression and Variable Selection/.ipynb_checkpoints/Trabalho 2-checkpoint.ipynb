## Correlações de maior ordem

Para analisar se ocorrem correlações de maior ordem significantes podemos elevar a coluna xi ao expoente analisado e verificar novamente o coeficiente de Pearson das variáveis, que nesse caso indicam essa relação não linear (pois xi está elevado ao esse novo expoente).

```python
df2 = df.copy()
col = f'x{i}^2'
df2[col] = df2[f'x{i}']**i
```

```python

corr_polinomiais = pd.DataFrame(index=[f'x^{j}' for j in range(1,11)])
for i in range(2,9):
    for exp in range(1,11):
        df2 = df.copy()
        index = f'x{i}^{exp}'
        df2[index] = df2[f'x{i}']**exp
        df_corr = df2.corr()
        df_corr = df2.corr()[[index]]
        df_corr = df_corr.drop(index)
        corr_polinomiais.at[f'x^{exp}', f'x{i}'] = df_corr[index]['y']
    
display(corr_polinomiais.style.background_gradient(cmap='RdBu',vmin=-1, vmax=1))
```

*Na tabela acima, o valor na linha x^j e coluna xi, indica o coeficiente de pearson entre os valores da coluna xi elevados a j e coluna y do dataframe principal.*

Como é possível ver na tabela acima, conforme também era possível inferir dos gráficos anteriores, não aparece nenhuma nova correlação, ou um correlação mais forte do que na linear (exceto em x6 que ocorre um aumento do coeficiente, porém mesmo assim ainda é um valor muito baixo para indicar correlação). Sendo assim, os termos de maior ordem não serão levados a diante na análise de seleção de variáveis.

#### P value regressão individual

Vamos inicialmente calcular a regressão simples para cada xi, e verificar o p value do coeficiente angular b1 e coeficiente linear b0:

```python
import statsmodels.api as sm

y = df['y'].to_numpy().reshape(-1,1)

table = pd.DataFrame(columns=['Y','X','b0_t','b1_t','b0_p','b1_p'])

regressions = {}

for col in df.drop('y',axis=1):
    x = df[col].to_numpy().reshape(-1,1)
    X = sm.add_constant(x)

    regression_model = sm.OLS(y, X).fit()

    y_predicted = regression_model.predict(X)

    b0_t = regression_model.tvalues[0]
    b1_t = regression_model.tvalues[1]

    b0_p = regression_model.pvalues[0]
    b1_p = regression_model.pvalues[1]

    table.loc[len(table)] = ['y', col, b0_t, b1_t, b0_p, b1_p]
    regressions[col] = regression_model
```

Nas próximas tabelas, b0_t e b1_t indicam o t value de b0 e b0, respectivamente, e b0_p e b1_p indicam o p value de b0 e b1, respectivamente.

```python
table.style.background_gradient(cmap='RdBu',vmin=-1, vmax=1, subset=['b0_p', 'b1_p'])
```

Como já era possível inferir das matrizes de correlações vistas anteriormente, x5 e x6 possuem p value muito alto para b1, então iremos descartar da análise essas variáveis.

```python
df = df.drop(['x5', 'x6'], axis=1)
```

## Correlações de interações

Podemos de forma análoga ao que foi feito anteriormente, verificar a correlação linear entre a variável y e as interações entre as variáveis:

```python
import itertools

df_interacao = df.copy()
df_interacao = df_interacao.drop('y', axis=1)

# Criar coluna xi*xj, para cada i != j
for pair in itertools.combinations(df_interacao.columns, 2):
    new_col = '*'.join(pair)
    df_interacao[new_col] = df_interacao[pair[0]] * df_interacao[pair[1]]

# Remover colunas que não são interações
df_interacao = df_interacao.drop(['x4','x7','x8'], axis = 1)

# Colocar a coluna y de volta para calcular a matriz de correlações
df_interacao['y'] = df['y']
```

```python
df_corr = df_interacao.corr()[['y']]
df_corr.drop('y').style.background_gradient(cmap='RdBu',vmin=-1, vmax=1)
```

Podemos ver que x7\*x8 possuem uma correlação significativa com y. Iremos então adicionar esse termo de interação no modelo.

```python
df['x7*x8'] = df['x7'] * df['x8'] 
```

```python
y = df['y'].to_numpy().reshape(-1,1)

regressions = {}
x = df.drop('y', axis=1).to_numpy()
X = sm.add_constant(x)

regression_model = sm.OLS(y, X).fit()

y_predicted = regression_model.predict(X)
```

```python
print(regression_model.summary())
```

Olhando o sumário acima, vemos que de fato temos o modelo prevê bem o comportamento da variável y, com um R² de 0.888, e com um p value baixo para todos os coeficientes do modelo.

Sendo assim, a regressão final fica:

$$\hat{y} = -1.1337\cdot x4 + 0.8638 \cdot x7 + 1.0103 \cdot x8 + 0.9475 \cdot x7 \cdot x8 + 1.3267$$

```python
df['y_hat'] = regression_model.fittedvalues
```
